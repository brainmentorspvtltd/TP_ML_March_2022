```python
columns = df.columns
```

```python
df_list = []
for i in range(len(x_df)):
    df_dict = {}
    for j in range(len(x_df[i])):
        df_dict[columns[j]] = x_df[i][j]
    df_list.append(df_dict)
```

```python
df_tuple = tuple(df_list)
```

```python
df_tuple[:5]
```

```python
model = compose.Pipeline(
    preprocessing.StandardScaler(),
    linear_model.LogisticRegression()
)
metric = metrics.Accuracy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_tuple, y, test_size=0.25)
```

```python
# x_train[:5]
```

```python
for features, target, test_feature, test_target in zip(x_train, y_train, x_test, y_test):
    model = model.fit_one(features, target)
    y_pred = model.predict_one(test_feature)
    metric = metric.update(test_target, y_pred)
print(metric)
```
