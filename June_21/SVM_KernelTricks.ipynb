```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import make_moons
```

```python
data, target = make_moons(n_samples=500)
```

```python
plt.scatter(data[:,0], data[:,1], c=target)
plt.show()
```

```python
def plotBoundary(k,c=1.0):
    plt.scatter(data[:,0], data[:,1], c=target)
    
    clf = SVC(kernel=k, C=c)
    clf.fit(data, target)
    
    # gca - get current axes
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    x1 = np.linspace(xlim[0], xlim[1], 200)
    x2 = np.linspace(ylim[0], ylim[1], 200)
    
    xx, yy = np.meshgrid(x1, x2)
    x = np.array([xx.flatten(), yy.flatten()]).T
    z = clf.decision_function(x).reshape(xx.shape)
    
    ax.contour(xx, yy, z, colors='k', levels=[-1,0,1], linestyles=['--', '-', '--'])
    ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=100, c='red')
    plt.show()
```

```python
plotBoundary('linear')
```

```python
plotBoundary('poly')
```

```python
plotBoundary('rbf')
```
