```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import make_moons
```

```python
data, target = make_moons(n_samples=500, noise=0.2)
```

```python
data[:5]
```

```python
plt.scatter(data[:,0], data[:,1], c=target)
plt.show()
```

```python
def plot_boundary(predict):
    x1 = np.arange(min(data[:,0]) - 1, max(data[:,0]) + 1, step=0.01)
    x2 = np.arange(min(data[:,1]) - 1, max(data[:,1]) + 1, step=0.01)
    
    xx, yy = np.meshgrid(x1, x2)
    x = np.array([xx.flatten(), yy.flatten()]).T
    z = predict(x)
    z = z.reshape(xx.shape)
    plt.contourf(xx,yy,z)
    plt.scatter(data[:,0], data[:,1], c=target, cmap=ListedColormap(('white','red')))
    plt.show()
```

```python
svm = SVC(kernel='rbf')
svm.fit(data, target)
```

```python
y_pred = svm.predict(data)
```

```python
accuracy_score(target, y_pred)
```

```python
plot_boundary(lambda x : svm.predict(x))
```
