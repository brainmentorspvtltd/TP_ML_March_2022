```python
import cv2
```

```python
img = cv2.imread('img.webp')
```

```python
img.shape
```

```python
cv2.imshow('result', img)
cv2.waitKey(0)
```

```python
# img
```

```python
img_slice = img[100:300, 100:300, :]
cv2.imshow('result', img_slice)
cv2.waitKey(0)
```

```python
# Resizing Image
```

```python
resize_img = cv2.resize(img, (300,400))
cv2.imshow('result', resize_img)
cv2.waitKey(0)
```

```python
resize_img_2 = cv2.resize(img, (300,360))
cv2.imshow('result', resize_img_2)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
```

```python
images = [img, img_slice, resize_img, resize_img_2]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
plt.show()
```

```python
# Blurring of Images
```

```python
# Gaussian Blur
# Median Blur
# Bilateral Filter
```

```python
gaussian = cv2.GaussianBlur(img, (7,7),20)
cv2.imshow('res', gaussian)
cv2.waitKey(0)
```

```python
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('res', grayImg)
cv2.waitKey(0)
```
